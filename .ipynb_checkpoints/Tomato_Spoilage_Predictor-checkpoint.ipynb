```python
import os
import pandas as pd

dataset_path = r"C:\Users\USER\Desktop\DSA\My Project\FGrade\data\Training_set"

# Updated mapping including '0'
label_to_days = {
    0: 1,
    1: 2,
    2: 3,
    3: 5,
    4: 7,
    5: 9,
    6: 11,
    7: 13,
    8: 15,
    9: 17
}

image_names = []
freshness_labels = []
approx_storage_days = []
spoilage_flags = []

for label_folder in os.listdir(dataset_path):
    if label_folder.isdigit():
        label = int(label_folder)
        if label in label_to_days:
            folder_path = os.path.join(dataset_path, label_folder)
            for img_file in os.listdir(folder_path):
                if img_file.lower().endswith(('.jpg', '.jpeg', '.png')):
                    image_names.append(img_file)
                    freshness_labels.append(label)
                    approx_storage_days.append(label_to_days[label])
                    spoilage_flags.append(1 if label >= 8 else 0)

df_fgrade = pd.DataFrame({
    'Image_Name': image_names,
    'Freshness_Label': freshness_labels,
    'Approx_Storage_Days': approx_storage_days,
    'Spoilage_Flag': spoilage_flags
})

print(df_fgrade.head())

df_fgrade.to_csv('fgrade_tomato_clean.csv', index=False)
print("✅ FGrade dataset cleaned and saved as 'fgrade_tomato_clean.csv'.")
```

```python

# Load FGrade dataset
df_fgrade = pd.read_csv('fgrade_tomato_clean.csv')

# Load Mendeley dataset
df_mendeley = pd.read_csv('mendeley_tomato_env.csv')

# Merge on 'Approx_Storage_Days' (FGrade) == 'Storage_Day' (Mendeley)
df_merged = pd.merge(
    df_fgrade,
    df_mendeley,
    left_on='Approx_Storage_Days',
    right_on='Storage_Day',
    how='left'
)

# Drop the duplicate 'Storage_Day' column after merging
df_merged = df_merged.drop(columns=['Storage_Day'])

# Save the merged dataset for model training
df_merged.to_csv('tomato_storage_training_data.csv', index=False)

print("✅ Merged dataset exported as 'tomato_storage_training_data.csv'.")
print(df_merged.head())
```

```python

# Train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Model training
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=42, class_weight='balanced')
model.fit(X_train, y_train)

# Evaluation
from sklearn.metrics import accuracy_score, classification_report
y_pred = model.predict(X_test)

print("✅ Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
importances = model.feature_importances_
features = X.columns

importance_df = pd.Series(importances, index=features).sort_values(ascending=True)
importance_df.plot(kind='barh')
plt.title("Feature Importances")
plt.show()
```
